## Magnetometry data analysis

### Aura Cuervo Gómez

```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib.mlab import griddata
```

#### Nota: En este ejercicio, los datos no están calibrados con la estación base del magnetometro

```python
def plottingdata(datos):
    x = datos[:,0]
    y = datos[:,1]
    b = datos[:,2]
    
    scatter(x,y,s=20,alpha=0.5)
    xlabel('Distancia en x')
    ylabel('Distancia en y')
    title('Grid')
    plt.savefig("Grid")
    
    fig = plt.figure()
    fig.set_size_inches(12.5,7.5)
    ax = fig.gca(projection='3d')

    xi = np.linspace(min(x),max(x))
    yi = np.linspace(min(y),max(y))

    X, Y = np.meshgrid(xi, yi)
    Z = griddata(x, y, b, xi, yi)

    surf = ax.plot_surface(X, Y, Z)

    ax.set_zlim3d(np.min(Z), np.max(Z))

    ax.set_xlabel('Eje x',fontsize=15)
    ax.set_ylabel('Eje y',fontsize=15)
    ax.set_zlabel('Magnetic field (nT)',fontsize=15)
    ax.set_title('Magnetic field',fontsize=25)
    plt.savefig("Magneto")
    plt.show()
```

### Group H

```python
datosH = np.loadtxt("MagnetoGrupoH.txt")
plottingdata(datosH)
```

#### En este caso, en la primera figura es claro que al tomar los datos el grupo no realizó una rejilla cuadrada. En cuanto a los valores del campo se observa un pico que hace que los demás valores parezcan muy homogeneos, sin embargo, al quitar el máximo se observa claramente la variación en el campo. Es muy díficil realizar conclusiones con respecto a la geología pues la confiabilidad de los datos es muy baja.

### Group E, line 2

#### Group E did not realize a grid in this line, so, it is not possible to generate a 3D plot

```python
datose = np.loadtxt("grupoe_line2.txt")

x = datose[:,0]
y = datose[:,1]
b = datose[:,2]

subplot(1,2,1)
scatter(x,y,s=20,alpha=0.5)
xlabel('Distancia en x')
ylabel('Distancia en y')
title('Grid')
plt.savefig("Grid")

subplot(1,2,2)
plt.plot(x, b, 'r')
xlabel('x')
ylabel('Magnetic field (nT)')
title('Magnetic field')
plt.savefig("MagnetoE_line2")
plt.show()
```
