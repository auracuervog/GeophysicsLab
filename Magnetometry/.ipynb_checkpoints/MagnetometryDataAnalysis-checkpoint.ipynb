### Magnetometry data analysis

#### Aura Cuervo Gómez

#### Group H

```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib.mlab import griddata
```

```python
#Loading the data collected in the field by group H

datosH = np.loadtxt("MagnetoGrupoH.txt")

xh = datosH[:,0]
yh = datosH[:,1]
bh = datosH[:,2]

#Number of points with magnetic field data
xmax = int(max(xh))
ymax = int(max(yh))
xi = (xmax/5)+1
yi = (ymax/5)+1

b = [[0 for x in xrange(xi)] for x in xrange(yi)] 

rows = 0
for line in datosH:
    rows+=1
    
for k in range(rows):
    if (int(xh[k]))==(int(xh[k-1])) & (int(yh[k]))==(int(yh[k-1])):
            meanfield = (bh[k]+bh[k-1])/2
            bh[k] = meanfield
            np.delete(xh,k-1)
            np.delete(yh,k-1)
            np.delete(bh,k-1)
        
for i in range(yi):
    for j in range(xi):
        b[i][j] = bh[i]
        
scatter(xh,yh,s=20,alpha=0.5)
xlabel('Distancia en x')
ylabel('Distancia en y')
```

```python
fig = plt.figure()
fig.set_size_inches(10.5,5.5)
ax = fig.gca(projection='3d')

xi = np.linspace(min(xh),max(xh))
yi = np.linspace(min(yh),max(yh))

X, Y = np.meshgrid(xi, yi)
Z = griddata(xh, yh, bh, xi, yi)

surf = ax.plot_surface(X, Y, Z)

ax.set_zlim3d(np.min(Z), np.max(Z))

ax.set_xlabel('Eje x',fontsize=15)
ax.set_ylabel('Eje y',fontsize=15)
ax.set_zlabel('Magnetic field (nT)',fontsize=15)
ax.set_title('Magnetic field',fontsize=25)

plt.show()
```

```python
maxb = np.max(bh)
posmax = bh.index(maxb)
np.delete(xh,posmax)
np.delete(yh,posmax)
np.delete(xh,posmax)

fig = plt.figure()
#fig.set_size_inches(30.5,15.5)
ax = fig.gca(projection='3d')

xi = np.linspace(np.min(xh),np.max(xh))
yi = np.linspace(np.min(yh),np.max(yh))


X, Y = np.meshgrid(xi, yi)
Z = griddata(xh, yh, bh, xi, yi)

surf = ax.plot_surface(X, Y, Z)

ax.set_zlim3d(np.min(Z), np.max(Z))       

ax.set_xlabel('Eje x',fontsize=18)
ax.set_ylabel('Eje y',fontsize=18)
ax.set_zlabel('Magnetic field (nT)',fontsize=18)
ax.set_title('Magnetic field',fontsize=40)

plt.show()
print posmax
```
