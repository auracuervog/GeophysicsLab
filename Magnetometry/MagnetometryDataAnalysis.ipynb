### Magnetometry data analysis

#### Aura Cuervo Gómez

#### Group H

```python
#Loading the data collected in the field by group H

datosH = np.loadtxt("MagnetoGrupoH.txt")

xh = datosH[:,0]
yh = datosH[:,1]
bh = datosH[:,2]

scatter(xh,yh,s=20,alpha=0.5)
xlabel('Distancia en x')
ylabel('Distancia en y')
```

```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib.mlab import griddata
```

```python
fig = plt.figure()
fig.set_size_inches(30.5,15.5)
ax = fig.gca(projection='3d')

xi = np.linspace(min(xh),max(xh))
yi = np.linspace(min(yh),max(yh))


X, Y = np.meshgrid(xi, yi)
Z = griddata(xh, yh, bh, xi, yi)

surf = ax.plot_surface(X, Y, Z)

ax.set_zlim3d(np.min(Z), np.max(Z))

ax.set_xlabel('Eje x',fontsize=18)
ax.set_ylabel('Eje y',fontsize=18)
ax.set_zlabel('Magnetic field (nT)',fontsize=18)
ax.set_title('Magnetic field',fontsize=40)

plt.show()
print len (bh)
```
